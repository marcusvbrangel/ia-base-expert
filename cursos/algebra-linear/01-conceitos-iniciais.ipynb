```python
# calculo matematico
# multiplicacao de cada um dos elementos por 2
tensor_np * 2
```

```python
# criaçao de tensor com a biblioteca tensorflow
tensor_tf = tf.convert_to_tensor(tensor_np)
```

```python
# linha x profundidade (nr. registros) x coluna -> 2 x 2 x 3
print(tensor_tf)
```

```python
type(tensor_tf)
```

```python
tensor_tf.ndim
```

```python
# calculo matematico
# multiplicacao de cada um dos elementos por 2
tensor_tf * 2
```

```python
# retornando o tensor no formato original do numpy
tensor_tf.numpy()
```

```python
# criaçao de tensor com a biblioteca pytorch
tensor_pt = torch.tensor(tensor_np)
```

```python
# calculo matematico
# multiplicacao de cada um dos elementos por 2
tensor_pt * 2
```

![SistemasLineares](./imagens/equacao-linear-01.jpeg)

---
Vamos plotar as retas referentes às formas simplificadas de `casal_1` e `casal_2`.

\begin{aligned}
127.35 &= x_1 + x_2\\
128.10 &= 2x_1 + 2x_2\\
\end{aligned}

<center>⏬ rearranjando para isolar $x_2$⏬</center>

\begin{aligned}
x_2 &= 127.35 - x_1\\
x_2 &= 64.05 - x_1
\end{aligned}

```python
# x1 sao os preços dos pratos pedidos
x1_preco_casal_1 = np.array([0, 10, 20, 30, 40, 50])
x1_preco_casal_2 = np.array([0, 10, 20, 30, 40, 50])
```

```python
# x2 sao a quantidade de pratos pedidos
x2_quantidade_casal_1 = 127.35 - x1_preco_casal_1
x2_quantidade_casal_2 = 64.05 - x1_preco_casal_2
```

```python
print(x2_quantidade_casal_1)
```

```python
print(x2_quantidade_casal_2)
```

```python
# plotar o grafico
plt.plot(x1_preco_casal_1, x2_quantidade_casal_1, label = "casal_1")
plt.plot(x1_preco_casal_2, x2_quantidade_casal_2, label = "casal_2")
plt.legend()
```

**Interpretação:**

As duas retas são paralelas, o que indica que este sistema não tem solução.

- Não há nenhuma quantidade de pratos que leve ao mesmo valor total.

---
Vamos avaliar um sistema formado por `casal_1` e `casal-3`.

\begin{aligned}
127.35 &= x_1 + x_2\\
134.85 &= 2x_1 + x_2\\
\end{aligned}

<center>⏬</center>

\begin{aligned}
x_2 &= 127.35 - x_1\\
x_2 &= 134.85 - 2x_1
\end{aligned}

```python
# x1 sao os preços dos pratos pedidos (casal_3)
x1_preco_casal_3 = np.array([0, 10, 20, 30, 40, 50])
```

```python
# x2 sao a quantidade de pratos pedidos (casal_3)
x2_quantidade_casal_3 = 134.85 - 2 * x1_preco_casal_3
```

```python
# plotar o grafico
plt.plot(x1_preco_casal_1, x2_quantidade_casal_1, label = "casal_1")
plt.plot(x1_preco_casal_3, x2_quantidade_casal_3, label = "casal_3")
plt.legend()
```

**Interpretação:**

As retas se cruzam, no ponto de encontro entre as duas retas no eixo "x", no valor aproximadamente de 9

e no eixo "y" aproximadamente 120...121, indica a solução do problema.
