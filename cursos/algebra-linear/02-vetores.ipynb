## Regularização

Efeito das normas L1 e L2 na regularização de algoritmos

Relação linear entre 3 variáveis independentes (x) e uma variável dependente (y)

![sdfsdf](./imagens/equacao-linear-02.jpeg)

```python
# especificação de valores aleatórios
coef_x1 = 2.5
coef_x2 = -1.8
coef_x3 = 0.05
intercept = 3.0
```

```python
# determinar os mesmos resultado para os numeros aleatorios
np.random.seed(10)
```

```python
# especificação da variáveis x1, x2 e x3
x1 = np.random.uniform(0, 10, size = 10)
x2 = np.random.uniform(0, 10, size = 10)
x3 = np.random.uniform(0, 10, size = 10)
```

```python
# exibição das variaveis geradas
x1, x2, x3
```

```python
# implementação da regressão linear (equação da reta y = ax + b)
# y é criado através da equação da reta: y = a1 * x1 + a2 * x2 + a3 * x3 + b
y = coef_x1 * x1 + coef_x2 * x2 + coef_x3 * x3 + intercept
```

```python
y
```

```python
# converter a representação das variáveis (x) para representação de matriz
X = np.column_stack((x1, x2, x3))
```

```python
X
```

```python
X.shape
```

![sdfsdf](./imagens/equacao-linear-03.jpeg)

#### Excutar uma regressao linear - sem regularização

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# exibir os novos coeficientes gerados
model.coef_
```

```python
# exibir o novo intercept
model.intercept_
```

**Interpretação**: 

Se observarmos temos os mesmos valors dos coeficientes, somente o valor do intercept que esta diferente, pois o intercept original
está arredondado.

Isto acontece pois estamos treinamento o algoritmo de regreção linear, nós estamos calculando a variável (y) como sendo uma função perfeitamente linear com os valores de (x), portanto temos os mesmos valores de coeficientes.

### Refazer o experimento usando Regressão do tipo Lasso (Norma L1)

![sdfsdf](./imagens/regressao-lasso-01.jpeg)

```python
from sklearn.linear_model import Lasso
```

```python
model_lasso = Lasso()
```

```python
model_lasso.fit(X, y)
```

```python
model_lasso.coef_
```

```python
model_lasso.intercept_
```

**Interpretação:**

Notamos agora que temos os valores dos coeficientes ligeiramente diferentes do original. Inclusive uns dos coeficientes (coef_x3) foi zerado.

Isto ocorreu porque a regressão linear do tipo lasso aplica a norma L1 para tentar zerar alguns coeficientes.

### Refazer o experimento usando Regressão do tipo Ridge (Norma L2)

```python
from sklearn.linear_model import Ridge
```

```python
model_ridge = Ridge()
```

```python
model_ridge.fit(X, y)
```

```python
model_ridge.coef_
```

```python
model_ridge.intercept_
```

**Interpretação:**

Nenhum coeficiente foi zerado, más o algoritmo tentou aproximar os coeficientes de zero o máximo possível.

**Valores Originais:**

- coef_x1 = 2.5
- coef_x2 = -1.8
- coef_x3 = 0.05
- intercept = 3.0

## Vetor unitário e vetor base

- Vetor Unitário: A Norma L2 é igual a 1, esta norma equivale ao comprimento do vetor.

![sdfsdf](./imagens/vetor-base-unitario-01.jpeg)

```python
# especificando que queremos obter a norma L2 (distancia euclidiana)
np.linalg.norm(vector, ord = 2)
```

```python
# vamos normalizar o vetor, dividindo ele por sua norma L2
vector_normalized = vector / np.linalg.norm(vector, ord = 2)
```

```python
vector_normalized
```

```python
# agora vamos verificar a norma do vetor normalizado, devemos ter o resultado igual a 1
# ou seja, este vetor normalizado pode ser considerado um vetor uniátario
np.linalg.norm(vector_normalized, ord = 2)
```

```python
# vetores base (i) e (j) com o valores de acordo com slide acima.
i = np.array([1, 0])
j = np.array([0, 1])
```

```python
7.5 * i + 8.25 * j
```

Repare que temos o mesmo resultado, pois o vetor unitário tem valor igual a 1.
